```python
import pandas as pd
```

```python
input_data = pd.read_csv(r"C:\Users\phani\OneDrive\Trainings\spam.csv",
                         encoding="latin",
                         usecols=["Target", "Text"])
input_data.head()
```

```python
input_data.shape
```

### Bag of Words Representation

```python
input_data["Text"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer(min_df=0.01, stop_words="english")
vec.fit(input_data["Text"])
transformed_data = vec.transform(input_data["Text"])
```

```python
transformed_data
```

```python
transformed_data = transformed_data.toarray()
print(vec.get_feature_names_out())
```

```python
transformed_data_df = pd.DataFrame(transformed_data, columns = vec.get_feature_names_out())
```

```python
transformed_data_df
```

### Model Building

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(transformed_data_df,input_data["Target"],test_size=0.33)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
```

```python
preds = clf.predict(x_test)
print(classification_report(y_test, preds))
```
